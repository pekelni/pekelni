По кількості строчок коду нечесний варіант програми не відрізняється (або майже не відрізняється) від чесного варіанту, але уся логіка повернута з ніг на голову: у чесній версії ми випадково визначаємо що обрали машини і далі йде нетривіально логіка по визначення переможця, а у цій, нечесній версії, ми випадково визначаємо переможця (проте ймовірність перемоги нечесна), а далі визначаємо що б мали обрати машини щоб наш вибір переможця справдився.

Наприклад, якщо користувач вибрав **камінь**, а ми випадковим (і нечесним) чином визначили що **машини** мають **перемогти**, то за допомогою `if-else` логіки ми визначаємо що єдиний елемент що переможе камінь - то є папір і далі робимо вигляд що машини випадковим чином вибрали папір.

Якщо користувач вибрав **камінь**, а ми випадковим (і нечесним) чином визначили що **машини** мають **програти** цього разу, то за допомогою `if-else` логіки ми визначаємо що єдиний елемент що програє каменю - то є ножиці і далі робимо вигляд що машини випадковим чином вибрали ножиці.
