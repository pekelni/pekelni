Ітеруватися по рядках просто: координата рядку (зазвичай `i`) не змінюється, натомість номер колонки пробігає усі значення від 0 до 7 (для РОЗМІР_ДОШКИ = 8).  

Ітеруватися по колонка просто: координата колонки (зазвичай `j`) не змінюється, натомість номер рядку пробігає усі значення від 0 до 7 (для РОЗМІР_ДОШКИ = 8).  

Щоб працювати з діагоналями треба мати змогу пройтися (ітеруватися) по усіх клітинках певної діагоналі. Подумайте, яким чином можна задати діагональ, яка формула описує координати діагоналі.

Для цього можете на аркуші паперу вибрати довільну діагональ (не забудьте про існування двох кутів нахилу діагоналі - "прямі" і "зворотні"), виписати координати клітинок (`i`, `j`) і придумати якій формулі вони відповідають. Формула дуже проста. Точніше там 2 прості формули для двох типів діагоналей.

Також подумайте (або порахуйте) скільки існує можливих діагоналей.
