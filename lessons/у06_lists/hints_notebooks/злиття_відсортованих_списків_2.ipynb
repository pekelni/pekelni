Назвемо два наші списки як список А і список Б, результат злиття - "результат". Далі нам необхідно ітеруватися по списку А і додавати елементи списку Б у результат допоки можливо. Що значить можливо? Це значить що у списку Б досі є недодані елементи і ці елементи менше поточного елемету зі списку А.

Далі ми виходимо із внутрішнього циклу і додаємо елемент зі списку А у результат (знов ж таки, тільки після того як ми додали усі менші елементи зі списку Б у результат).

Зовнішній цикл - це буде ітерація по елементах списку А. Внутрішня - ітерація по деяким елементам списку Б (допоки у списку Б є елементи менші за поточний елемент списку А).
