Для кожної кульки можете моделювати усі її зіткнення випадковим чином і виводити послідовність типу `'01011'`, де 1 означає падіння вправо, 0 - вліво. Далі задача буде як конвертувати таку послідовність у номер слоту в який врешті решт потрапить кулька.

Переконайтесь, що ви обрали правильну кількість цвяхів для заданої кількості слотів.

Також корисне питання: на яку відстань (горизонтально) зміщується кулька коли падає на цвях і "повертає" вправо чи вліво? Скільки таких падінь відбувається для кожної кульки? Яка початкова координата кульки (по горизонталі)?
