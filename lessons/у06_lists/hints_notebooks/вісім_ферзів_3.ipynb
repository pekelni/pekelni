Ось код, який випадковим чином генерує розташування ферзів що частково задовільняють умові задачі - не б'ють один одного вздовж рядків і колонок, але ви можете бачете що у кожному варіанті з цих 3 випадковів варіантів ферзі знаходяться на однакових діагоналях.

Вам треба змінити код таким чином, щоб він генерував такі випадкові варіанти допоки всі ферзі не будуть на різних діагоналях. Тобто кожен варіант ви маєте перевіряти кожну діагональ щоб побачити скільки ферзів там знаходиться, і якщо умова задачі порушується пропускати такий варіант.

Попередня задача про генерацію усіх діагоналей має вам тут допомогти.

```python
import random

РОЗМІР_ДОШКИ = 8
дошка = list()
ПУСТИЙ_РЯДОК = list()
for j in range(РОЗМІР_ДОШКИ):
    ПУСТИЙ_РЯДОК.append(False)
for i in range(РОЗМІР_ДОШКИ):
    рядок = list(ПУСТИЙ_РЯДОК)
    рядок[i] = True
    дошка.append(рядок)

for i in range(3):
    random.shuffle(дошка)

    for рядок in дошка:
        print(end='|')
        for клітина in рядок:
            символ = 'Q' if клітина else ' '
            print(символ, end='|')
        print()
    print()
```
