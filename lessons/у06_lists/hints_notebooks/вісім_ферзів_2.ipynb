Для початку можете написати програму яка просто генерує кілька випадкових розташувань ферзів, наприкалд 10.

Створіть двовимірний список 8 на 8 який представляє дошку. Далі заповніть усю головну діагональ ферзями в одному циклі, тобто позиції `(0, 0), (1, 1), (2, 2), ...`. Тепер ви можете випадковим чином перемішувати рядки на такій дошці і ферзі ГАРАНТОВАНО не зможуть атакувати один одного вздовж рядків чи колонок, але багато варіантів буде таких де ферзі можуть один одного атакувати по діагоналях, ці варіанти доведеться пропускати допоки не знайдете хороший варіант.
