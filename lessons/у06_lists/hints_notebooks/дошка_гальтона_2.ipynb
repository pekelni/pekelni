Якщо розглядати ілюстрацію, то бачимо що для `8` слотів у нас вбито `7` рядів цвяхів, тобто кулька буде вдарятися `7` разів. Також просто з картинки можно побачити що від удару кулька зміщується на половину слота, тобто `0.5` або `-0.5`.

На початку падіння кулька розміщується рівно МІЖ третім та четвертим слотами (якщо рахувати з нуля), тобто початкова координана буде дорівнювати `3.5`.

Розглянемо кілька прикладів:
- кулька впада `7` разів вправо. На початку `слот = 3.5` (між третім на четвертим). За `7` падінь вправо кулька змістилася на `0.5 * 7`, тобто потрапила у слот `3.5 + 3.5 = 7`
- кулька впада 7 разів вліво. На початку `слот = 3.5` (між третім на четвертим). За `7` падінь вліво кулька змістилася на `-0.5 * 7`, тобто потрапила у слот `3.5 - 3.5 = 0`
- кулька впада `5` разів вліво і `2` рази вправо. На початку `слот = 3.5` (між третім на четвертим). За `5` падінь вліво кулька змістилася на `-0.5 * 5`, за `2` падіння вправо - на `0.5 * 2`, тобто потрапила у слот `3.5 - 2.5 + 1 = 2`
