Ви можете використати `time.time()` для того, щоб визначити поточний час у секундах. Далі вам необхідно конвертувати його у години. У цій задачі нас не цікавлять секунди чи дробні частини хвилини, тому застосуйте відповідні конвертації чи округлення після того як переведете час у хвилини.

На момент написання цієї підказки пройшло 452300 годин з 1 січня 1970. Подумайте як таку велику кількість годин можна конвертувати у кількість годин "всередені" доби, тобто у число від 0 до 23 (бо зазвичай ми не бачимо на годинниках час типу 452300:28).

Нічого страшного якщо вам це не інтуітивно, ми вивчили не так багато арифметичних операцій (десь 6), можете просто спробувати яка підійде.

Для першої спроби можете не використовувати узагалі часовий пояс, а просто надрукувати поточний час як він є, виходячи зі значення що повернеться із `time.time()`, це буде час по Грінвічу.
