## Вступ
Змінна - іменована комірка в пам’яті що зберігає певне значення (наприклад число, шматок тексту, список покупок, дату народження). Значення змінної можна змінювати та зчитувати. Спершу ми розглянемо як працюють змінні, потім розберемо чим вони корисні.

Змінну можна задати і потім використати для виводу на екран:

````python
змінна = 'значення змінної;'
інша_змінна = 'а це інше значення;'
числова_змінна = 1234
ffffffffffff = 99
x = 7
print(змінна, інша_змінна, числова_змінна)
print(ffffffffffff, x)
````

Назва змінної може складатися з букв, цифр і нижніх прочерків і не може починатися із цифри чи містити пробіли.

Змінну можна зчитувати у арифметичних виразах:

````python
а = 1
б = 2
print((а + б) * б ** б + 100500)
````

У змінну можна записувати результати операцій, змінну можна перевизначати:

````python
а = 3
б = а + 3
а = б - 10
print(а, б)
````

Якщо ви перевизначаєте змінну багато разів і потім зчитуєте, то використовується останнє значення

````python
а = 1
а = 2
print(а)
а = 3
а = 4
print(а)
````

## Позбуваємося магічних чисел
На цьому етапі вам може здатися що змінна - це такий складний спосіб записати числа чи рядки і що можна просто напряму писати числа і не морочити собі голову. Розглянемо приклад який частково показує користь змінних.

Що ви думаєте про наступне обчислення?

````python
print(6.67e-11 * 5.972e24 * 7.35e22 / 384_400_000 ** 2)
````

Навряд чи ви зрозумієте що там рахується, якщо тільки не захоплюєтеся фізикою та астрономією.

А тепер використаємо змінні:

````python
гравітаційна_стала = 6.67e-11
маса_землі = 5.972e24
маса_місяця = 7.35e22
відстань = 384_400_000
сила_гравітації = гравітаційна_стала * маса_землі * маса_місяця / відстань ** 2
print(сила_гравітації)
````

Тут ви скорше за все одразу зрозумієте що проводиться розрахунок сили гравітації між Землею і Місяцем навіть якщо ви вперше в житті бачите формулу закону гравітації.

Пряме використання чисел у коді, без зрозуміло названих змінних, називається **магічними числами** (magic numbers) і це антипаттерн до якого схильні початківці.

До речі, порівняйте тепер формулу у програмерському стилі із формулою у науковому стилі:

````python
g = 6.67e-11
mz = 5.972e24
mm = 7.35e22
r = 384_400_000
f = g * mz * mm / r ** 2
print(f)

# де g це гравітаційна стала
# де mz це маса землі
# де mm це маса місяця
# де r це відстань
# де f це сила_гравітаціїи
````

Наукова нотація була започаткована пару століть тому, коли економити букви мало сенс.

Код пишеться один раз, а читається потім багато разів. До того ж автодоповнення у Пайчармі (PyCharm) робить використання навіть довгих імен безболісним.

Моя порада - давайте зрозумілі назви вашим змінним якщо ви пишете щось довше ніж 3 рядки коду.

Я у цьому курсі використовую назви змінних на українській мові для полегшення сприйняття, але мушу сказати що постійно переключати мову з англійської на українську і навпаки незручно, тому у самостійній роботі я раджу використовувати транслітерацію (`masa_zemli` замість `маса_землі`) або англійську мову (`earth_mass` замість `маса_землі`).

## Літерали
"Числа, написані напряму" називаються літералами. Точніше визначення: літерал - **постійне значення (певного типу даних), записане у (програмному) коді**. Рядки теж можуть бути записані у вигляді літералів. Насправді ми вже знайомі з літералами, зараз же вчимо як це правильно називається.

У наступному коді `назва_змінної` і `назва_іншої_змінної` то, раптово, назви змінних, а `1234` та `"привіт"` - літерали.

````python
назва_змінної = 1234
назва_іншої_змінної = "привіт"
print(назва_змінної, назва_іншої_змінної)
````

Відмінності літерала від змінної:
- у літерала завжди одне і те ж значення, яке ви можете побачити не запускаючи програму
- у літерала немає імені, лише значення
- у літерал не можна нічого присвоїти

Схожості літерала і змінної:
- і те і те можна передавати у функції
- і те і те можна використовувати для обчислень (за умови доречного типу даних)

Проілюструємо "у літерал не можна нічого присвоїти":

````python
a = 1
1 = a
````

У наступному прикладі бачимо що і літерал, і змінну можна передати у функцію `type()`.

````python
a = 1
print(type(a))
print(type(1))
````

Зверніть увагу що значок `=` означає **присвоєння**, а не просто "дорівнює" як у математиці.

Про присвоєння треба думати як однонаправлену операцію, присвоєння можна інтуітивніше зобразити стрілочкою:
```
a ← 1  # правильно
1 ← a  # неправильно, неможливо присвоювати у літерал
```
`=` простіше набирати з клавіатури і всі звикли до `=`, тому використовується `=` у якості оператору присвоєння.

Тепер, знаючи основи змінних, ми можемо краще зрозуміти чому рядки у Пайтоні необхідно записувати у лапках. Без лапок Пайтон може "сприймати" рядки як назви змінних і скаржитись якщо така змінна не була задана.

Пам'ятайте, що якщо ви **не** задали змінну то ви не можете її використовувати: виникне помилка імені (`NameError`).

````python
print("привіт")
print(привіт)
````

````python
привіт = 1234
print('привіт')
print(привіт)
````

Функція `input()` дозволяє приймати вхідні дані від користувача.

## Функція `random.randint()`

Ця функція дозволяє генерувати випадкове ціле число. Функції які ми розглядали до цього - `input()`, `int()`, `print()`, `float()`, `str()` (деякі з них за сумісництвом є і типами) є вбудованими (**built-in**). Це означає що їх можна завжди використовувати без імпорту. Усього у Пайтоні приблизно 70 вбудованих функцій.

Якщо ми спробуємо використати `random.randint()` без імпорту то матимемо помилку:

````python
випадкове_число = random.randint(1, 666)
print(випадкове_число)
````

`random.randint()` знаходиться у модулі `random`. Модуль це, грубо кажучи, збірник функцій (і класів і деяких інших конструкцій). Коли програміст пише велику програму (скажімо, на 1000 рядків коду і більше), то стає доречним розбивати функціонал на модулі задля кращого логічного групування, приблизно як конспект зазвичай розбивається на окремі лекції, а книга - на розділи.

````python
import random
````

````python
випадкове_число = random.randint(1, 666)
print(випадкове_число)
````

````python
print(random.randint(1, 555), random.randint(1, 555))
print(random.randint(1, 555), random.randint(1, 555))
print(random.randint(1, 555), random.randint(1, 555))
print(random.randint(1, 555), random.randint(1, 555))
````

Тепер ми можемо написати спрощений тренер арифметики. На кожен виклик він буде генерувати випадкову вправу на додавання.

````python
import random

доданок_0 = random.randint(1, 50)
доданок_1 = random.randint(1, 50)
правильна_відповідь = доданок_0 + доданок_1
питання_для_користувача = str(доданок_0) + ' + ' + str(доданок_1) + ' = '
відповідь_користувача = int(input(питання_для_користувача)) 
print('Ваша відповідь: ' + str(відповідь_користувача) + ', правильна відповідь: ' + str(правильна_відповідь))
````

## Функція `time.time()`
Ця функція видає кількість секунд, які пройшли на момент виклику з першого січня 1970 року, 00:00:00 у UTC часовій зоні. UTC це universal coordinated time - всесвітній координований час. Раніше його неформально називали "час по Грінвічу".

````python
import time

print(time.time())
````

Цю функцію можна використати щоб замірити час який займає якесь складне обчислення. У веб-програмуванні зазвичай немає таких складних обчислень, але, наприклад, звернення до бази даних може займати значні долі секунди (0.01 - 0.5 секунд), а якщо бути необережним або викачувати багато інформації то запит може зайняти декілька секунд і більше.

У таких випадках корисно заміряти який код зайняв скільки часу щоб зрозуміти які саме шматки коду доведеться покращувати.

````python
import time

початковий_час = time.time()
результат_розрахунку = 666_666 ** 4_444
тривалість = time.time() - початковий_час
print('Простий розрахунок зайняв ' + str(тривалість) + ' секунд')

початковий_час = time.time()
результат_розрахунку = 666_666 ** 666_666
тривалість = time.time() - початковий_час
print('Складніший розрахунок зайняв ' + str(тривалість) + ' секунд')
````

## Необхідність змінних
Ми щойно розглянули два важливі приклади - спрощений тренер арифметики і замір тривалості обчислення, які добре ілюструють ще одну причину існування змінних: робота зі значеннями невідомими на момент написання програми.

У тренері арифметики ми не знаємо наперед який приклад буде згенеровано і яку відповідь введе користувач. Відповідно неможливо для представлення таких значень використати літерали, лише змінні дозволять нам представити таку інформацію, що ще не існує на момент написання програми.

Те саме із заміром часу: ми не знаємо наперед коли саме користувач запустить програму, відповідно результат виклику `time.time()` буде постійно зростати. Також ми не знаємо тривалість обчислення до тих пір поки його не заміряємо, до того ж є певна варіабельність у цій тривалості.

Це і є головною причиною існування змінних: **робота зі значеннями невідомими наперед**.

У великих проектах ми, наприклад, ніколи не знаємо наперед що саме нам поверне база даних (які самі товари в онлайн магазині чи яких саме пацієнтів у електронній системі лікарні), або що введе користувач у веб-формі для оформлення замовлення.

## Функція `round()`
Функція `round()` дозволяє округляти числа, наприклад:

````python
print(round(12.1111111111111111111111111))
````

Проте того ж самого ефекту можна досягти за допомогою конвертації у `int`:

````python
print(int(12.1111111111111111111111111))
````

Але функція `round()` стане у нагоді якщо ви хочете округлити число до, наприклад, двох знаків після коми:

````python
print(round(12.111111111111111111111111, 2))
````

Значення аргумента можна змінювати і округлювати до різної кількості знаків:

````python
pi = 3.14159265359
print(round(pi, 1))
print(round(pi, 2))
print(round(pi, 3))
print(round(pi, 4))
````

Один із варіантів використання цієї функції це коли ви хочете приховати зайві деталі від користувача бо вони йому не цікаві. Уявіть, наприклад, що в онлайн магазині ви побачити щось накшталт:

````python
вартість = 1234  # гривні
знижка = 16.66666666  # відсотки
вартість_зі_знижкою = вартість * (1 - знижка / 100)
print('Фінальна вартість з усіма знижками: ' + str(вартість_зі_знижкою) + ' гривень')
````

Це б виглядало дуже дивно, округлення вирішує цю проблему:

````python
print('Фінальна вартість з усіма знижками: ' + str(round(вартість_зі_знижкою, 2)) + ' гривень')
````

Варто зазначити, що у справжніх проект такое округлення зазвичай досягається за допомогою форматування рядків:

````python
print('Фінальна вартість з усіма знижками: {:.2f} гривень'.format(вартість_зі_знижкою))
````

Але як бачите, це виглядає досить страшно і незрозуміло, тому вивчати такий функціонал початківцям недоречно. Я досі не вивчив на пам'ять `{:.2f}` не зважаючи на роки досвіду і часто гуглю як воно робиться.

## Модифіковане присвоєння (augmented assignment)
Часто виникають ситуації коли певну змінну треба збільшити на певне значення:

````python
змінна = 1
доданок = 5
print(змінна)
змінна = змінна + 3
змінна = змінна + доданок
print(змінна)
````

Є коротша форма написання цієї самої операції:

````python
змінна = 1
доданок = 5
print(змінна)
змінна += 3
змінна += доданок
print(змінна)
````

Таким чином `змінна += 3` означає: "Пайтон, збільши змінну `змінна` на `3`".

`змінна += доданок` означає: "Пайтон, збільши змінну `змінна` на `доданок`".

Модифіковане присвоєння працює і для інших операторів: `- * / // % **`.

````python
змінна = 10
змінна -= 1  # те саме що: змінна = змінна - 1
print(змінна)
````

````python
змінна = 10
змінна *= .5  # те саме що: змінна = змінна * .5
print(змінна)
````

````python
змінна = 10
змінна /= 3  # те саме що: змінна = змінна / 3
print(змінна)
````

````python
змінна = 10
змінна //= 3  # те саме що: змінна = змінна // 3
print(змінна)
````

````python
змінна = 10
змінна %= 3  # те саме що: змінна = змінна % 3
print(змінна)
````

````python
змінна = 10
змінна **= 3  # те саме що: змінна = змінна ** 3
print(змінна)
````

## Накопичення (агрегація) інформації
Часто у коді необхідно зібрати інформацію із багатьох об'єктів (значень) в одну змінну. Наприклад, якщо ми хочемо порахувати вартість усіх товарів у корзині онлайн-магазину чи порахувати суму квадратів усіх чисел від 0 до 10.

Для цього часто використовується одна змінна значення якої ми постійно міняємо додаючи нові значення. Зазвичай такий код використовує цикли і списки, але ми ознайомимося із цими структурами пізніше, тому у цьому прикладі я їми не користуюся.

````python
сума = 0
сума += 1 ** 2
сума += 2 ** 2
сума += 3 ** 2
сума += 4 ** 2
сума += 5 ** 2
сума += 6 ** 2
сума += 7 ** 2
сума += 8 ** 2
сума += 9 ** 2
print(сума)
````

## Підсумок
- **змінні** - іменовані комірки у пам'яті. Змінні можна змінювати, у змінні можна присвоювати різні значення багато разів
- **присвоєння** - операція запису певного значення у змінну, це одностороння, направлена операція, а не операція "дорівнює" як у математиці. Операція присвоєння більш доречно позначалася б однонаправленою стрілкою
- **літерал** - постійне значення, записане у коді. Літерали незмінні, не можна щось присвоїти у літерал, отримаєте `SyntaxError`. Рядкові літерали записуються у лапках, числові - без лапок.
- щоб користуватися змінною, її необхідно спершу визначити, інакше отримаєте `NameError`
- Головна причина існування змінних - робота зі значеннями, які невідомі наперед, на момент написання програми. Наприклад: ввід користувача, випадкові числа, тривалість обчислень, список товарів що були видані базою даних, інформація про пацієнта який 5 хвилин тому був прийнятий у кімнаті швидкої допомоги.
- Додаткові застосування змінних: боротьба із магічними числами (покращування читабельності), використання одного значення багато разів, накопичення інформації в одній змінній.
- Функція `input()` дозволяє зчитати ввід користувача. `input()` приймає параметр-підказку що пояснить користувачу що саме ми хочемо щоб він ввів. `input()` завжди повертає рядок, навіть якщо користувач вводить цифри.
- Деякі функції перед використанням необхідно імпортувати, наприклад `random.randint()`, інакше отримаєте `NameError`
- `random.randint()` приймає 2 параметри і повертає випадкове ціле число, кожен раз - нове (але з певної ймовірністю числа можуть співпадати). `random.random()` не приймайє аргументів і повертає випадкове число від 0 до 1
- функція `time.time()` видає кількість секунд що пройшли з 1 січня 1970
- функція `len()` приймає один аргумент (наприклад рядок) і повертає його довжину
- для виклику функції необхідно використовувати дужки, інакше Пайтон сприйме функцію просто як назву змінної
- **модифіковане присвоєння** дозволяє більш лаконічно збільшувати значення змінної на певне число. Модифіковане присвоєння працює і з іншими математичними операціями
- У шкільній математиці здебільшого використовуються невідомі і константи які не змінюються у процесі розв’язку певної задачі, змінні ж у програмуванні часто змінються по мірі виконання програми
