Ось приклад циклу який друкує усі необхідні комбінації (і трохи більше). Вам необхідно відсіяти ті де є трійка і додати логіку яка друкує числа і робить перенос рядку після кожного десятого числа.

```python
for цифра_1 in range(1, 5):
    for цифра_2 in range(5):
        for цифра_3 in range(5):
            print(цифра_1, цифра_2, цифра_3)
```
